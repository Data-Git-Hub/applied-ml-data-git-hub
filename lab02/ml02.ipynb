### Introduction
The RMS Titanic disaster of April 15, 1912, is one of the most infamous maritime tragedies in history. With an estimate 2,200 passengers and crew on board, the ship's collision with an iceberg led to the loss of more than 1,500 lives. The tragedy was shaped by multiple factors, including social class divisions, lifeboat shortages, and survival prioritization based on gender, age, and ticket class. The Titanic dataset, compiled from historical records, provides valuable insights into these dynamics, making it a widely used resource for statistical analysis and machine learning applications. <br>

The Titanic dataset from Seaborn contains demographic and logistical details of the passengers and crew, offering a structured look at factors that influenced survival rates. Key features include passenger class `pclass`, gender `sex`, age `age`, number of siblings/spouses aboard `sibsp`, number of parents/children aboard `parch`, fare price `fare`, and embarkation point `embarked`. The target variable, survived, indicates whether a passenger lived or perished. Analyzing this dataset allows for a deeper understanding of survival patterns, class disparities, and decision-making processes during the disaster. <br>

Exploring this dataset reveals significant patterns and anomalies. First-class passengers had a higher survival rate due to preferential treatment in lifeboat access, while third-class passengers faced greater challenges escaping the lower decks. Women and children were more likely to survive, reinforcing the "women and children first" policy. However, missing values in age and embarkation data introduce challenges that require careful data cleaning and handling. By leveraging visualization and predictive modeling, we can identify key survival determinants and build models that assess the likelihood of survival based on passenger characteristics. <br>

2.1.B. Generate one Boxenplot for Each Column (Good for Large Datasets) <br>

Next, Boxenplots. <br>


```python
# Generate a list of unique colors (same as histograms)
numeric_columns = titanic_data.select_dtypes(include=['number']).columns  # Select only numeric columns
colors = sns.color_palette("husl", n_colors=len(numeric_columns))  # Assign distinct colors

# Set figure size for multiple Boxen plots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  # Adjust grid layout based on available numeric features
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Loop through each numeric feature and create a Boxen plot
for i, col in enumerate(numeric_columns):
    sns.boxenplot(y=titanic_data[col], ax=axes[i], color=colors[i])  # Assign matching colors
    axes[i].set_title(f"Boxen Plot: {col}")  # Add title for each feature

# Hide any unused subplot spaces
for i in range(len(numeric_columns), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()

```


2.1.C. Generate All Scatter Plots<br>

Third, Scatter Plots. <br>


```python
# Generate a list of unique colors (same as histograms and boxen plots)
numeric_columns = titanic_data.select_dtypes(include=['number']).columns  # Select only numeric columns
colors = sns.color_palette("husl", n_colors=len(numeric_columns))  # Assign distinct colors

# Create scatter plots for all feature pairs
feature_combinations = list(itertools.combinations(numeric_columns, 2))  # Get all unique column pairs
num_plots = len(feature_combinations)

# Determine grid size (3 columns per row, adjust rows automatically)
num_cols = 3
num_rows = (num_plots // num_cols) + (num_plots % num_cols > 0)  # Ensure enough rows

# Set figure size dynamically
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, num_rows * 4))
axes = axes.flatten()  # Flatten the array for easy iteration

# Loop through each feature pair and create a scatter plot
for i, (col_x, col_y) in enumerate(feature_combinations):
    color_x = colors[list(numeric_columns).index(col_x)]
    sns.scatterplot(x=titanic_data[col_x], y=titanic_data[col_y], ax=axes[i], color=color_x, alpha=0.6)

    # Set titles and labels
    axes[i].set_title(f"{col_x} vs {col_y}")
    axes[i].set_xlabel(col_x)
    axes[i].set_ylabel(col_y)

# Hide any extra empty subplots
for i in range(len(feature_combinations), len(axes)):
    fig.delaxes(axes[i])  # Removes unused subplot spaces

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

### Section 4. Train a Linear Regression Model

4.1 Split the data <br>

Split the dataset into training and test sets (80% train / 20% test). <br>

```python
from titanic import train_test_split

# Split the data into training and test sets (80% train / 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    df_X, df_y, test_size=0.2, random_state=42
)

# Display confirmation
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

4.2 Train the model <br>

Create and fit a `LinearRegression` model. <br>

LinearRegression – A class from sklearn.linear_model that creates a linear regression model. <br>

model – An instance of the LinearRegression model. This object will store the learned coefficients and intercept after training. <br>

```python
# Create a Linear Regression model instance
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, y_train)

# Display confirmation
print("Model training completed successfully!")
```

4.2.A Make Predictions for the Test Set. <br>

The model.predict() method applies this equation to the X test data to compute predicted values. <br>

y_pred = model.predict(X_test) <br>

y_pred contains all the predicted values for all the rows in X_test based on the linear regression model. <br> 

```python
# Generate predictions using the test set
y_pred = model.predict(X_test)

# Display confirmation and a preview of predictions
print("Predictions for the test set completed successfully!\n")

# Show a comparison of actual vs. predicted values (first 10 rows)
comparison_df = pd.DataFrame({"Actual": y_test.values, "Predicted": y_pred})
display(comparison_df.head(10))
```

4.3 Report R^2, MAE, RMSE <br>

Evaluate the model using R^2, MAE, and RMSE. <br>

4.3.A. First: R^2<br>

- Coefficient of Determination (R^2) - This tells you how well the model explains the variation in the target variable. A value close to 1 means the model fits the data well; a value close to 0 means the model doesn’t explain the variation well. <br>

```python
# Calculate R² score
r2 = r2_score(y_test, y_pred)

# Display result
print(f'R² Score: {r2:.2f}')
```

4.3.B. Second: MAE<br>

- Mean Absolute Error (MAE) - This is the average of the absolute differences between the predicted values and the actual values. A smaller value means the model’s predictions are closer to the actual values. <br>

```python
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Display result
print(f'Mean Absolute Error (MAE): {mae:.2f}')
```

4.3.C. Third: RMSE <br>

- Root Mean Squared Error (RMSE) - This is the square root of the average of the squared differences between the predicted values and the actual values. It gives a sense of how far the predictions are from the actual values, with larger errors having more impact. <br>

```python
# Calculate Root Mean Squared Error (RMSE) manually
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Display result
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
```
